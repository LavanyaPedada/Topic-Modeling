```python
import glob
import os

os.chdir("/data1/call_miner/Vinyas_Call_miner/call_miner")
#file_list = glob.glob(os.path.join(os.getcwd(), "Feb_Text_data", "*.txt"))
file_list = glob.glob(os.path.join(os.getcwd(), "Feb_Cust_words", "*.txt"))

corpus = []

for file_path in file_list:
    with open(file_path) as f_input:
        corpus.append(f_input.read())
```

```python
import re
# Remove Emails
%time corpus = [re.sub('\S*@\S*\s?', '', sent) for sent in corpus]

# Remove new line characters
%time corpus = [re.sub('\s+', ' ', sent) for sent in corpus]

# Remove distracting single quotes
%time corpus = [re.sub("\'", "", sent) for sent in corpus]
```

```python
print(corpus[2])
```

```python
print(summarize(str(corpus[2]),ratio=0.1))
```

```python
from gensim.summarization import keywords

keywords(str(corpus[2]),ratio=1,scores=True,pos_filter=('NN', 'JJ','VB'),lemmatize=True,deacc=True)
```
