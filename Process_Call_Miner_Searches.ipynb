Read and process callminer searches to unique word list

## Processing Call Miner searches

```python
import glob
import os

os.chdir("/data1/call_miner/Vinyas_Call_miner/call_miner")
#file_list = glob.glob(os.path.join(os.getcwd(), "Feb_Text_data", "*.txt"))
file_list = '/data1/call_miner/Vinyas_Call_miner/call_miner/dispute_searches.txt'
data=[]

with open(file_list) as f_input:
    data.append(f_input.read())
```

```python
data
```

```python
import re

data = [x.replace("|", " ") for x in data]
data = [x.replace('"', "") for x in data]
data = [x.replace('(', "") for x in data]
```

```python
data = [x.replace(')', "") for x in data]
data = [re.sub('\s+', ' ', x) for x in data]
data = [x.replace(']', "") for x in data]
data = [x.replace('[', "") for x in data]
data = [re.sub("\`", "", x) for x in data]
```

```python
import re

data = [re.sub('(\d{1})(\w{1})',"",num) for num in data] # replace reversed7w
```

```python
data = [re.sub("\'", "", x) for x in data]
```

```python
import gensim
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))  # deacc=True removes punctuations

%time data_tokens = list(sent_to_words(data))
```

```python
data_tok = [y for x in data_tokens for y in x]
```

```python
data_tok
```

```python
unique_tok=set(data_tok)
```

```python
print "Total words:",len(data_tok)
print "Unique words:",len(unique_tok)
```
